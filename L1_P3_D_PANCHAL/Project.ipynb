```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

#### We are adding a feature 'has_reviewed' which will be a flag indicating that the properity has ever been reviewed or not

```python
df['has_reviewed'] = (df['number_of_reviews'] > 0).astype(int)
```

#### As we can see that the 'price' attribute is highly Right skewed so we need to do log transform first

#### For Scaling purpose we use the RoustScaler because it utilizes median and IQR, as the StandardScaler uses mean to scale, which in this case would be high due to the max value

```python
df['price'] = np.log1p(df['price'])
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

cols = [
    'latitude',
    'longitude',
    'price',
    'minimum_nights',
    'number_of_reviews',
    'reviews_per_month',
    'calculated_host_listings_count',
    'availability_365'
]

df[cols] = scaler.fit_transform(df[cols])
```

```python
df.to_csv("AB_NYC_2019_cleaned_scaled.csv", index=False)
```
